<h1>10.5. Объектно-ориентированный подход</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./data/tips.csv',header=0)
df.head()
```

<h2>Объекно-ориентированный подход</h2>

```python
# Настроки
fig = plt.figure()
# Плоскость
axes = fig.add_axes([0,0,1,1])
# bins увеличивает количество столбцов, color - цвет
axes.hist(df.total_bill,bins=20,color='green')
# Устанавливаем заголовок диаграммы
axes.set_title('Общая сумма счета')
# заголовок оси ординат
axes.set_ylabel('Количество заказов')
# заголовок для оси абсцисс
axes.set_xlabel('Сумма заказа, USD')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.scatter(x=df.total_bill,y=df.tip,marker=',')
axes.set_title('Общая сумма счета Vs Сумма чаевых')
axes.set_xlabel('Сумма чека')
axes.set_ylabel('Размер чаевых')
```
