<h1>10.10 Гистограммы распределения признаков</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Предупреждения
#import warnings;
#warnings.simplefilter('ignore')

df = pd.read_csv('./data/tips.csv')
df.head()
```

<h2>Гистограмма распределения признаков</h2>
<p>Для построения <b>гистограммы распределения количественных признаков</b> в библиотеке Seaborn можно использовать метод <b style="color:#00FF00">distplot</b></p>

```python
sns.set()
sns.distplot(df.total_bill)
```

<p>Seaborn позволяет разместить на одном графике гистограммы распределения двух признаков и показать связь между ними</p>

```python
sns.jointplot(x='total_bill',y='tip',data=df)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
sns.pairplot(df)
```
