<h1>10.8 Subplots</h1>
<p>Разместить в одном пространстве несколько графиков</p>

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./data/tips.csv')
df.head()
```

```python
fig,axes = plt.subplots(1,3,figsize=(12,4))

axes[0].hist(df.total_bill)
axes[0].set_title('Общая сумма чаевых')

axes[1].hist(df.tip)
axes[1].set_title('Сумма чаевых')

axes[2].scatter(x=df.total_bill,y=df.tip,marker='o')
axes[2].set_title('Общая сумма счета Vs Сумма чаевых')
```
