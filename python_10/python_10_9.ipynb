<h1>10.9 Создание линейных графиков</h1>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(start=-3.,stop=3.,num=1000)
y1 = np.exp(x)
```

<h2>Отрисовка одной кривой</h2>

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y1)
```

<h2>Отрисовка нескольких кривых</h2>

```python
y2 = 10 + 0.4*x - 0.3*x**2 + 0.1*x**3

fig = plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x[500:],y1[500:])
axes.plot(x,y2)
```

<h2>Стили линий</h2>

```python
fig = plt.figure(figsize=(7,4))
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y1,linestyle='--')
```

<h2>Цвета и маркеры</h2>

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x[::30],y2[::30],'mD')
```
