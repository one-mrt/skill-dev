<h1>10.4. Отображение двух показателей на графике</h1>


```python
import pandas as pd

df = pd.read_csv('./data/tips.csv',header=0)
df.head()
```

<h2>Отображение двух показателей на графике</h2>
<p>Метод plot позволяет использовать <b>параметр subplots</b>, который принимает логические значения True или False. При установке значения True можно разместить несколько графиков в одной рабочей области</p>

```python
df[['total_bill','tip']].plot(kind='hist',
                            grid=True,
                            subplots=True,
                            title=['Общие сумма счета', 'Сумма чаевых'],
                            legend=False)
```

```python
df.plot(x='total_bill',
        y='tip',
        kind='scatter',
        grid=True,
        title='Общая сумма счета Vs сумма чаевых')
```

```python
df.pivot_table(values=['total_bill','tip'],
                index='day',
                aggfunc='mean').plot(kind='bar')
```

<h2>Задание 1 </h2>
<p>Какой заголовок лучше подойдет для диаграммы, созданной с помощью этого кода</p>

```python
df.pivot_table(values='tip',
                index=['sex','smoker'],
                aggfunc='count').plot(kind='bar')
```

<h2>Задание 2</h2>
<p>Какой вариант кода лучше</p>

```python
df.groupby('smoker').tip.mean().plot(kind='barh')
```

```python
df.smoker.value_counts().plot(kind='bar')
```

```python
df.groupby('smoker').tip.count().plot(kind='bar')
```
