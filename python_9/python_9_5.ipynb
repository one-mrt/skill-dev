## 9.5. **Математические и статистические операции**

```python
students = np.array([
                [1,135,34,4],
                [2,160,43,5],
                [3,163,40,4.3],
                [4,147,44,5],
                [5,138,41,4.7],
                [6,149,54,3.9],
                [7,136,39,4.2],
                [8,154,48,4.9],
                [9,137,35,3.7],
                [10,165,60,4.6],
            ],dtype=float)

students
```

### **Средние величины**
**Среднее арифметическое (mean)** - сумма всех значений, деленная на их количество, показывает общую тенденцию данных и описывает их одним числом.

```python
mean = np.mean(students[:,-1])
mean
```

**Медиана** - это такое число, что равно половина элементов из выборки больше него, а другая половина меньше него.
Если в выборке есть выбросы (значения, которые принимаю величину существенно выше или ниже седнего и выделяются из всей выборки),
то медиана будет лучше характеризовать всю выборку, чем среднее арифметическое

```python
median = np.median(students[:,-1])
print(f'Медиана {median}')

sort=np.sort(students[:,-1])
(sort[4]+sort[5])/2
```

### **Задание 2**

### **Коэффициент корреляции**
Корреляция - статистическая взаимосвязь случайных величин.
Мерой корреляции служит одноименный коэффициент, который показывает, насколько сильно связаны величины, он может быть положительным или отрицательным, а по модулю принимает значение от 0 до 1. Отрицательный коэффициент говорит о том, что случайные величины связаны, но при увеличении одной из них вторая уменьшается. Если коэффициент положительный, то величины изменяются в одном направлении.

```python
corr = np.corrcoef(students[:,1],students[:,2])
corr
```

## **Задание 3**
Между какой парой признаков в массиве students наблюдается минимальная корреляция, из предложенных вариантов

```python
corGrowMas = np.corrcoef(students[:,1], students[:,2])
corGrowPer = np.corrcoef(students[:,1],students[:,3])
corMasPer = np.corrcoef(students[:,2], students[:,3])

print('Между ростом и массой',corGrowMas)
print('Между ростом и успеваемостью',corGrowPer)
print('Между массой тела и успеваемостью',corMasPer)
```

<h2><b>Показатели вариации</b></h2>
<p><b>Математическое ожидание</b> - среднее значение случайной величины при стремлении числа ее измерений к <b>бесконечности</b>.
Если число измерений конечно, то для оценки среднего значения величины используется среднее арифметическое</p>
<h3>
<b>Дисперсия</b> - мера разброса случайной величины относительно ее математического ожидания.
</h3>
<p>
<b>D[X] = M[(X - M[X])<sup>2</sup>]</b>
</p>
<p class="f-16px">Дисперсия случайно величины Х равна математическому ожиданию квадрата отклонения случайной величины от ее математического ожидания. То есть, если величина дисперсии небольшая, значит, все числа в выборке имеют близкие друг к другу значения, а чем она больше - тем значительнее разброс показателей</p>

<p><b>Стандартное отклонение</b> - самый распространенный показатель рассеивания случайной величины относительно ее математического ожидания</p>
<p><b>q = ^D[X]</b></p>
<p>Стандартное отклонение равно квадратному корню из дисперсии. Низкое стандартное отклонение показывает, что все значения в выборке сгруппированы около среднего значения. Большой показатель этой величины говорит о том, что разброс значений большой</p>

```python
std = np.std(students[:,1])
std
```
